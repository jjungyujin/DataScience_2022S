## 데이터 표준화

표준화를 통해 변수의 범위, scale, unit을 통일

```python
df=pd.read_csv('data/auto.csv')
df
```

```python
# 표준화
df=(df-df.mean())/df.std()
```

## SKlearn을 이용한 전처리
SKlearn : 파이썬에서 표준화된 머신러닝 라이브러리

```python
df=pd.read_csv('data/auto.csv')
df
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
# 스케일러 객체 생성
# StandardScaler : 평균 0, 표준편차 0으로 스케일링
scaler = StandardScaler()
output = scaler.fit_transform(df) # numpy array 형태로 반환
output
```

```python
pd.DataFrame(output, columns=df.columns)
```

## Polynomial Term 추가하기

선형회귀 모형의 한계를 깨기 위한 방법 (비선형적인 관계 표현하기)
- 다차항 넣기
- 변수간의 상호작용

```python
# 독립변수 추리기
df=pd.read_csv('data/classification_data.csv')
x = df[['X1', 'X2']]
x
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2, include_bias=False)
output = poly.fit_transform(x)
```

```python
c = poly.get_feature_names(x.columns)
c
```

```python
pd.DataFrame(output, columns=c)
```

```python
# 차수 변경
poly = PolynomialFeatures(degree=3, include_bias=False)
new_output = poly.fit_transform(x)
new_output
```

```python
new_c = poly.get_feature_names(x.columns)
new_c
```

```python
pd.DataFrame(new_output, columns = new_c)
```
